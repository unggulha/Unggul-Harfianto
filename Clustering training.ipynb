```python
kmeans=KMeans(n_clusters=3).fit(y)
centroids=kmeans.cluster_centers_
centroids=np.array(centroids)
centroids
```

```python
C1=centroids[0,:]
C2=centroids[1,:]
C3=centroids[2,:]
```

```python
# calculate the Euclidean diatance between the data 

def calculate_distance(centroid, X, Y):
    distances = []
        
    # Unpack the x and y coordinates of the centroid
    c_x, c_y = centroid
        
    # Iterate over the data points and calculate the distance
    for x, y in list(zip(X, Y)):
        root_diff_x = (x - c_x) ** 2
        root_diff_y = (y - c_y) ** 2
        distance = np.sqrt(root_diff_x + root_diff_y)
        distances.append(distance)
    
    return distances
```

```python
# Calculate the distance and assign them to the DataFrame accordingly

y['C1_Distance'] = calculate_distance(C1, y.potential, y.overall)
y['C2_Distance'] = calculate_distance(C2, y.potential, y.overall)
y['C3_Distance'] = calculate_distance(C3, y.potential, y.overall)

# Preview the data
print(y)
```

```python
# Get the minimum distance centroids
y['Cluster'] = y[['C1_Distance', 'C2_Distance', 'C3_Distance']].apply(np.argmin, axis =1)
    
# Map the centroids accordingly and rename them
y['Cluster'] = y['Cluster'].map({'C1_Distance': 'C1', 'C2_Distance': 'C2', 'C3_Distance': 'C3'})
    
# Get a preview of the data
print(y)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
feature=pd.concat([fifa.age,fifa.weight_kg,fifa.height_cm,fifa.overall,fifa.potential,fifa.value_eur,fifa.wage_eur],axis=1)
feature
```

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(feature)
```

```python
#Calculate covariance matrix
cov_mat = np.cov(X_std.T)

#Calculate covariance matrix (in addition)
#mean_vec = np.mean(X_std, axis=0)
#cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0]-1)

#Perform eigen decomposition on covariance matrix
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print('\nEigenvalues \n%s' %eig_vals)
print('Eigenvectors \n%s' %eig_vecs)
```

```python
#Calculate Eigen Using Sklearn (As Performed Above)
pca=PCA(n_components=7)
X_pca = pca.fit(X_std).transform(X_std)
print('\nEigenvalues \n%s' %pca.explained_variance_)
print('Eigenvectors \n%s' %pca.components_)
```

```python
#Print Scree Plot
eigen_val = pca.explained_variance_
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.title('Scree Plot of PCA: Component Eigen Values')
plt.plot(eigen_val)
print(eigen_val)
```

```python
#cumulative variance explains (proportion)
var1=np.cumsum(np.round(pca.explained_variance_ratio_,decimals=4)*100)
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Variance')
plt.title('Explained Variance of PCA by Component')
plt.plot(var1)
print(var1)
```
